# **1. Introdução**



* **Planejamento** é o raciocínio ao lado da ação.
    * Escolhe e organiza ações pela antecipação dos seus resultados esperados.
    * **Planejamento Automatizado:** área da IA que estuda este processo automaticamente.
* Várias formas de planejamento em razão da existência de diversos tipos de ações.
* **Planejamento específico de domínio**: compreende o endereçamento de cada problema com representações específicas e técnicas adpatadas ao problema.
    * Há alguns problemas:
        * Semelhanças entre os problemas não são endereçadas nesta abordagem. O estudo destas semelhanças pode contribuir na melhoria da abordagem.
        * Custoso para endereçar cada problema de planejamento ao invés de confiar e adptar algumas ferramentas gerais.
        * Uma abordagem de domínio específico torna dificultosa a projeção de sistemas autônomos inteligentes. Limitação por áreas.
* O Planejamento Automatizado está interessado em uma *abordagem independente de domínio*.
    * Na resolução de um dado problema, o planejador (nesta abordagem) tem como entrada as especificações do problema e o conhecimento sobre o domínio.


> *Automated Planning is not meant to be opposed to domain-specific planning techniques, just as automated reasoning is not intended to replace every arithmetic and float-point calculus processor and other specialized reasoning techniques in a computer.* (GHALLAB, NAU E TRAVERSO, 2004, p. 4)

* O propósito do planejamento automatizado é desenvolver *abordagens gerais para uma forma particular de pensamento* que pode ser *construída* e pode ser *bem integrada com ferramentas de planejamento de domínio específico*.

* Modelo gerais e abstração de ações são elementos com os quais o planejamento independente de domínio conta.

## **1.1. Modelo Conceitual para o Planejamento**

* O planjeamento se interessa em escolher e organizar ações para mudar o estado de um sistema.
* Modelagem baseada em um *sistema de transição de estados*: $\Sigma = (S, A, E, \gamma)$
    * $S = ${$s_1, s_2, \dots$} é um conjunto finito ou recursivamente enumerável de estados.
    * $A = ${$a_1, a_2, \dots$} é um conjunto finito ou recursivamente enumerável de ações.
    * $E = ${$e_1, e_2, \dots$} é um conjunto finito ou recursivamente enumerável de estados.
    * $\gamma: S \times A \times E \rightarrow 2^S$ é uma função de transição de estados.
